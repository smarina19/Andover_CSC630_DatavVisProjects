```python
import pandas as pd
df = pd.read_csv('wealth_share_data.csv', ';')
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

df.plot(x='Year',y=['Top 1% Share','Top 10% Share', 'Bottom 50% Share'],color=['Red','Green','Blue'], ylabel = 'Share of Wealth',title='Share of Wealth owned by Top 1%, Top 10%,\n and Bottom 50% in the US since 1962')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.35), ncol= 2)
plt.show()
```
